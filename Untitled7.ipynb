```python
import tensorflow as tf
from tensorflow.keras import layers, models
import pathlib
from google.colab import drive
drive.mount('/content/drive')

def build_resnet101_model(input_shape=(224, 224, 3), num_classes=2):
    input_tensor = layers.Input(shape=input_shape)

    # Initial Convolutional Layer
    conv1 = layers.Conv2D(64, (7, 7), strides=(2, 2), padding='same')(input_tensor)
    bn1 = layers.BatchNormalization()(conv1)
    relu1 = layers.Activation('relu')(bn1)
    maxpool1 = layers.MaxPooling2D((3, 3), strides=(2, 2))(relu1)

    def residual_block(x, filters, stride=1, conv_shortcut=False):
        shortcut = x
        if conv_shortcut:
            shortcut = layers.Conv2D(filters * 4, (1, 1), strides=stride, padding='same')(shortcut)
            shortcut = layers.BatchNormalization()(shortcut)

        conv2a = layers.Conv2D(filters, (1, 1), strides=stride, padding='same')(x)
        bn2a = layers.BatchNormalization()(conv2a)
        relu2a = layers.Activation('relu')(bn2a)

        conv2b = layers.Conv2D(filters, (3, 3), padding='same')(relu2a)
        bn2b = layers.BatchNormalization()(conv2b)
        relu2b = layers.Activation('relu')(bn2b)

        conv2c = layers.Conv2D(filters * 4, (1, 1), padding='same')(relu2b)
        bn2c = layers.BatchNormalization()(conv2c)

        add = layers.add([bn2c, shortcut])
        output = layers.Activation('relu')(add)
        return output

    # ResNet Blocks - using a list to represent the block structure
    resnet_blocks = [
        (64, 3, 1),  # filters, repetitions, initial stride
        (128, 4, 2),
        (256, 23, 2),
        (512, 3, 2)
    ]

    x = maxpool1
    for filters, repetitions, stride in resnet_blocks:
        x = residual_block(x, filters, stride=stride, conv_shortcut=True)
        for _ in range(repetitions - 1):
            x = residual_block(x, filters)

    # Final Layers
    avgpool = layers.AveragePooling2D((7, 7))(x)
    flatten = layers.Flatten()(avgpool)
    dense1 = layers.Dense(512, activation='relu')(flatten)
    output_tensor = layers.Dense(num_classes, activation='softmax')(dense1)

    model = models.Model(inputs=input_tensor, outputs=output_tensor)
    return model

# --- Data Loading using image_dataset_from_directory ---
data_directory = pathlib.Path('/content/drive/MyDrive/Project/ORAL CANCER')  # Ensure this is the correct path
image_dimension = (224, 224)
batch_size_value = 32
training_dataset = tf.keras.utils.image_dataset_from_directory(
    data_directory,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=image_dimension,
    batch_size=batch_size_value,
    label_mode='categorical',
    labels='inferred',
    interpolation='nearest',
    shuffle=True
)

validation_dataset = tf.keras.utils.image_dataset_from_directory(
    data_directory,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=image_dimension,
    batch_size=batch_size_value,
    label_mode='categorical',
    labels='inferred',
    interpolation='nearest',
    shuffle=False
)

class_labels = training_dataset.class_names
number_of_classes = len(class_labels)
print(f"Class labels: {class_labels}")
print(f"Number of classes: {number_of_classes}")

# Instantiate the model with the determined number of classes
resnet101_model = build_resnet101_model(num_classes=number_of_classes)

# Compile the model
resnet101_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Model Summary
resnet101_model.summary()

# Train the model
number_of_epochs = 10
training_history = resnet101_model.fit(training_dataset, validation_data=validation_dataset, epochs=number_of_epochs)
```
